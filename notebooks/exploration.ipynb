```python
# %% [markdown]
# # House Price Prediction - Data Exploration
# 
# Exploration initiale du dataset pour comprendre la structure des données et identifier les patterns.

# %%
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path
import warnings
warnings.filterwarnings('ignore')

# Configuration des graphiques
plt.style.use('seaborn-v0_8')
sns.set_palette("husl")
plt.rcParams['figure.figsize'] = (12, 8)

# %% [markdown]
# ## 1. Chargement des données

# %%
# Chemin vers les données
data_path = Path("../data/raw/data.csv")

# Chargement du dataset
try:
    df = pd.read_csv(data_path)
    print(f"✅ Dataset chargé avec succès: {df.shape[0]} lignes, {df.shape[1]} colonnes")
except FileNotFoundError:
    print("❌ Fichier non trouvé. Téléchargez le dataset depuis:")
    print("https://www.kaggle.com/datasets/shree1992/housedata?select=output.csv")
    print("Et placez-le dans data/raw/output.csv")
    df = None

# %%
# Aperçu des données
if df is not None:
    print("📊 Aperçu des premières lignes:")
    display(df.head())
    
    print("\n📋 Informations générales:")
    df.info()
    
    print("\n📈 Statistiques descriptives:")
    display(df.describe())

# %% [markdown]
# ## 2. Analyse des valeurs manquantes

# %%
if df is not None:
    # Calcul des valeurs manquantes
    missing_data = df.isnull().sum()
    missing_percent = (missing_data / len(df)) * 100
    
    missing_df = pd.DataFrame({
        'Colonnes': missing_data.index,
        'Valeurs manquantes': missing_data.values,
        'Pourcentage': missing_percent.values
    }).sort_values('Pourcentage', ascending=False)
    
    print("🔍 Analyse des valeurs manquantes:")
    display(missing_df[missing_df['Valeurs manquantes'] > 0])
    
    # Visualisation
    if missing_df['Valeurs manquantes'].sum() > 0:
        plt.figure(figsize=(12, 6))
        missing_cols = missing_df[missing_df['Valeurs manquantes'] > 0]
        sns.barplot(data=missing_cols, x='Pourcentage', y='Colonnes')
        plt.title('Pourcentage de valeurs manquantes par colonne')
        plt.xlabel('Pourcentage (%)')
        plt.tight_layout()
        plt.show()
    else:
        print("✅ Aucune valeur manquante détectée!")

# %% [markdown]
# ## 3. Analyse de la variable cible

# %%
if df is not None:
    # Identifier la colonne de prix (souvent 'price', 'Price', 'target', etc.)
    price_cols = [col for col in df.columns if 'price' in col.lower()]
    
    if price_cols:
        target_col = price_cols[0]
        print(f"🎯 Variable cible identifiée: {target_col}")
        
        # Statistiques de base
        print(f"\n📊 Statistiques du prix:")
        print(f"Minimum: ${df[target_col].min():,.2f}")
        print(f"Maximum: ${df[target_col].max():,.2f}")
        print(f"Moyenne: ${df[target_col].mean():,.2f}")
        print(f"Médiane: ${df[target_col].median():,.2f}")
        print(f"Écart-type: ${df[target_col].std():,.2f}")
        
        # Distribution des prix
        fig, axes = plt.subplots(1, 3, figsize=(18, 5))
        
        # Histogramme
        axes[0].hist(df[target_col], bins=50, alpha=0.7, color='skyblue', edgecolor='black')
        axes[0].set_title('Distribution des prix')
        axes[0].set_xlabel('Prix ($)')
        axes[0].set_ylabel('Fréquence')
        
        # Box plot
        axes[1].boxplot(df[target_col])
        axes[1].set_title('Box plot des prix')
        axes[1].set_ylabel('Prix ($)')
        
        # Log transformation
        log_prices = np.log1p(df[target_col])
        axes[2].hist(log_prices, bins=50, alpha=0.7, color='lightcoral', edgecolor='black')
        axes[2].set_title('Distribution des prix (log)')
        axes[2].set_xlabel('Log(Prix + 1)')
        axes[2].set_ylabel('Fréquence')
        
        plt.tight_layout()
        plt.show()
        
        # Test de normalité
        from scipy import stats
        _, p_value = stats.normaltest(df[target_col])
        print(f"Test de normalité (p-value): {p_value:.6f}")
        if p_value < 0.05:
            print("❌ Distribution non normale (transformation recommandée)")
        else:
            print("✅ Distribution approximativement normale")
    
    else:
        print("⚠️ Colonne de prix non trouvée automatiquement")
        print("Colonnes disponibles:", df.columns.tolist())

# %% [markdown]
# ## 4. Analyse des variables numériques

# %%
if df is not None:
    # Sélection des colonnes numériques
    numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()
    print(f"🔢 Variables numériques ({len(numeric_cols)}): {numeric_cols}")
    
    if len(numeric_cols) > 1:
        # Matrice de corrélation
        correlation_matrix = df[numeric_cols].corr()
        
        plt.figure(figsize=(14, 12))
        mask = np.triu(np.ones_like(correlation_matrix, dtype=bool))
        sns.heatmap(correlation_matrix, mask=mask, annot=True, 
                   cmap='coolwarm', center=0, fmt='.2f',
                   square=True, cbar_kws={"shrink": .8})
        plt.title('Matrice de corrélation des variables numériques')
        plt.tight_layout()
        plt.show()
        
        # Variables les plus corrélées avec le prix
        if price_cols:
            target_col = price_cols[0]
            correlations = df[numeric_cols].corr()[target_col].abs().sort_values(ascending=False)
            print(f"\n🔗 Corrélations avec {target_col}:")
            for var, corr in correlations.items():
                if var != target_col:
                    print(f"{var}: {corr:.3f}")

# %% [markdown]
# ## 5. Analyse des variables catégorielles

# %%
if df is not None:
    # Sélection des colonnes catégorielles
    categorical_cols = df.select_dtypes(include=['object', 'category']).columns.tolist()
    print(f"📝 Variables catégorielles ({len(categorical_cols)}): {categorical_cols}")
    
    for col in categorical_cols[:5]:  # Limiter aux 5 premières
        print(f"\n📊 {col}:")
        value_counts = df[col].value_counts()
        print(f"  Valeurs uniques: {df[col].nunique()}")
        print(f"  Valeurs les plus fréquentes:")
        print(value_counts.head())
        
        # Graphique pour les variables avec peu de catégories
        if df[col].nunique() <= 10:
            plt.figure(figsize=(10, 6))
            value_counts.plot(kind='bar', color='lightgreen', alpha=0.8)
            plt.title(f'Distribution de {col}')
            plt.xlabel(col)
            plt.ylabel('Fréquence')
            plt.xticks(rotation=45)
            plt.tight_layout()
            plt.show()

# %% [markdown]
# ## 6. Détection des valeurs aberrantes

# %%
if df is not None and len(numeric_cols) > 0:
    print("🎯 Détection des valeurs aberrantes (méthode IQR):")
    
    outliers_summary = []
    
    for col in numeric_cols[:6]:  # Limiter aux 6 premières colonnes numériques
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        
        outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
        outlier_count = len(outliers)
        outlier_percent = (outlier_count / len(df)) * 100
        
        outliers_summary.append({
            'Variable': col,
            'Nombre': outlier_count,
            'Pourcentage': outlier_percent,
            'Borne inf': lower_bound,
            'Borne sup': upper_bound
        })
        
    outliers_df = pd.DataFrame(outliers_summary)
    display(outliers_df)

# %% [markdown]
# ## 7. Résumé et recommandations

# %%
if df is not None:
    print("📋 RÉSUMÉ DE L'EXPLORATION")
    print("=" * 50)
    print(f"📊 Taille du dataset: {df.shape[0]:,} lignes × {df.shape[1]} colonnes")
    print(f"🔢 Variables numériques: {len(numeric_cols)}")
    print(f"📝 Variables catégorielles: {len(categorical_cols)}")
    
    missing_total = df.isnull().sum().sum()
    print(f"❓ Valeurs manquantes: {missing_total:,} ({(missing_total/(df.size))*100:.2f}%)")
    
    if price_cols:
        target_col = price_cols[0]
        print(f"🎯 Variable cible: {target_col}")
        print(f"💰 Prix moyen: ${df[target_col].mean():,.2f}")
    
    print("\n🔧 RECOMMANDATIONS POUR LE PREPROCESSING:")
    print("=" * 50)
    
    # Recommandations basées sur l'exploration
    if missing_total > 0:
        print("• Gérer les valeurs manquantes (imputation ou suppression)")
    
    if len(categorical_cols) > 0:
        print("• Encoder les variables catégorielles (One-Hot, Label)")
        
    if price_cols:
        target_col = price_cols[0]
        skewness = df[target_col].skew()
        if abs(skewness) > 1:
            print(f"• Transformer la variable cible (skewness: {skewness:.2f})")
    
    print("• Détecter et traiter les valeurs aberrantes")
    print("• Normaliser/standardiser les variables numériques")
    print("• Créer de nouvelles features si pertinent")

# %%
print("✅ Exploration terminée!")
print("📁 Prochaine étape: Preprocessing des données")
```
