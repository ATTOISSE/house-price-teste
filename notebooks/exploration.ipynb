```python
# %% [markdown]
# # House Price Prediction - Data Exploration
# 
# Exploration initiale du dataset pour comprendre la structure des donn√©es et identifier les patterns.

# %%
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path
import warnings
warnings.filterwarnings('ignore')

# Configuration des graphiques
plt.style.use('seaborn-v0_8')
sns.set_palette("husl")
plt.rcParams['figure.figsize'] = (12, 8)

# %% [markdown]
# ## 1. Chargement des donn√©es

# %%
# Chemin vers les donn√©es
data_path = Path("../data/raw/data.csv")

# Chargement du dataset
try:
    df = pd.read_csv(data_path)
    print(f"‚úÖ Dataset charg√© avec succ√®s: {df.shape[0]} lignes, {df.shape[1]} colonnes")
except FileNotFoundError:
    print("‚ùå Fichier non trouv√©. T√©l√©chargez le dataset depuis:")
    print("https://www.kaggle.com/datasets/shree1992/housedata?select=output.csv")
    print("Et placez-le dans data/raw/output.csv")
    df = None

# %%
# Aper√ßu des donn√©es
if df is not None:
    print("üìä Aper√ßu des premi√®res lignes:")
    display(df.head())
    
    print("\nüìã Informations g√©n√©rales:")
    df.info()
    
    print("\nüìà Statistiques descriptives:")
    display(df.describe())

# %% [markdown]
# ## 2. Analyse des valeurs manquantes

# %%
if df is not None:
    # Calcul des valeurs manquantes
    missing_data = df.isnull().sum()
    missing_percent = (missing_data / len(df)) * 100
    
    missing_df = pd.DataFrame({
        'Colonnes': missing_data.index,
        'Valeurs manquantes': missing_data.values,
        'Pourcentage': missing_percent.values
    }).sort_values('Pourcentage', ascending=False)
    
    print("üîç Analyse des valeurs manquantes:")
    display(missing_df[missing_df['Valeurs manquantes'] > 0])
    
    # Visualisation
    if missing_df['Valeurs manquantes'].sum() > 0:
        plt.figure(figsize=(12, 6))
        missing_cols = missing_df[missing_df['Valeurs manquantes'] > 0]
        sns.barplot(data=missing_cols, x='Pourcentage', y='Colonnes')
        plt.title('Pourcentage de valeurs manquantes par colonne')
        plt.xlabel('Pourcentage (%)')
        plt.tight_layout()
        plt.show()
    else:
        print("‚úÖ Aucune valeur manquante d√©tect√©e!")

# %% [markdown]
# ## 3. Analyse de la variable cible

# %%
if df is not None:
    # Identifier la colonne de prix (souvent 'price', 'Price', 'target', etc.)
    price_cols = [col for col in df.columns if 'price' in col.lower()]
    
    if price_cols:
        target_col = price_cols[0]
        print(f"üéØ Variable cible identifi√©e: {target_col}")
        
        # Statistiques de base
        print(f"\nüìä Statistiques du prix:")
        print(f"Minimum: ${df[target_col].min():,.2f}")
        print(f"Maximum: ${df[target_col].max():,.2f}")
        print(f"Moyenne: ${df[target_col].mean():,.2f}")
        print(f"M√©diane: ${df[target_col].median():,.2f}")
        print(f"√âcart-type: ${df[target_col].std():,.2f}")
        
        # Distribution des prix
        fig, axes = plt.subplots(1, 3, figsize=(18, 5))
        
        # Histogramme
        axes[0].hist(df[target_col], bins=50, alpha=0.7, color='skyblue', edgecolor='black')
        axes[0].set_title('Distribution des prix')
        axes[0].set_xlabel('Prix ($)')
        axes[0].set_ylabel('Fr√©quence')
        
        # Box plot
        axes[1].boxplot(df[target_col])
        axes[1].set_title('Box plot des prix')
        axes[1].set_ylabel('Prix ($)')
        
        # Log transformation
        log_prices = np.log1p(df[target_col])
        axes[2].hist(log_prices, bins=50, alpha=0.7, color='lightcoral', edgecolor='black')
        axes[2].set_title('Distribution des prix (log)')
        axes[2].set_xlabel('Log(Prix + 1)')
        axes[2].set_ylabel('Fr√©quence')
        
        plt.tight_layout()
        plt.show()
        
        # Test de normalit√©
        from scipy import stats
        _, p_value = stats.normaltest(df[target_col])
        print(f"Test de normalit√© (p-value): {p_value:.6f}")
        if p_value < 0.05:
            print("‚ùå Distribution non normale (transformation recommand√©e)")
        else:
            print("‚úÖ Distribution approximativement normale")
    
    else:
        print("‚ö†Ô∏è Colonne de prix non trouv√©e automatiquement")
        print("Colonnes disponibles:", df.columns.tolist())

# %% [markdown]
# ## 4. Analyse des variables num√©riques

# %%
if df is not None:
    # S√©lection des colonnes num√©riques
    numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()
    print(f"üî¢ Variables num√©riques ({len(numeric_cols)}): {numeric_cols}")
    
    if len(numeric_cols) > 1:
        # Matrice de corr√©lation
        correlation_matrix = df[numeric_cols].corr()
        
        plt.figure(figsize=(14, 12))
        mask = np.triu(np.ones_like(correlation_matrix, dtype=bool))
        sns.heatmap(correlation_matrix, mask=mask, annot=True, 
                   cmap='coolwarm', center=0, fmt='.2f',
                   square=True, cbar_kws={"shrink": .8})
        plt.title('Matrice de corr√©lation des variables num√©riques')
        plt.tight_layout()
        plt.show()
        
        # Variables les plus corr√©l√©es avec le prix
        if price_cols:
            target_col = price_cols[0]
            correlations = df[numeric_cols].corr()[target_col].abs().sort_values(ascending=False)
            print(f"\nüîó Corr√©lations avec {target_col}:")
            for var, corr in correlations.items():
                if var != target_col:
                    print(f"{var}: {corr:.3f}")

# %% [markdown]
# ## 5. Analyse des variables cat√©gorielles

# %%
if df is not None:
    # S√©lection des colonnes cat√©gorielles
    categorical_cols = df.select_dtypes(include=['object', 'category']).columns.tolist()
    print(f"üìù Variables cat√©gorielles ({len(categorical_cols)}): {categorical_cols}")
    
    for col in categorical_cols[:5]:  # Limiter aux 5 premi√®res
        print(f"\nüìä {col}:")
        value_counts = df[col].value_counts()
        print(f"  Valeurs uniques: {df[col].nunique()}")
        print(f"  Valeurs les plus fr√©quentes:")
        print(value_counts.head())
        
        # Graphique pour les variables avec peu de cat√©gories
        if df[col].nunique() <= 10:
            plt.figure(figsize=(10, 6))
            value_counts.plot(kind='bar', color='lightgreen', alpha=0.8)
            plt.title(f'Distribution de {col}')
            plt.xlabel(col)
            plt.ylabel('Fr√©quence')
            plt.xticks(rotation=45)
            plt.tight_layout()
            plt.show()

# %% [markdown]
# ## 6. D√©tection des valeurs aberrantes

# %%
if df is not None and len(numeric_cols) > 0:
    print("üéØ D√©tection des valeurs aberrantes (m√©thode IQR):")
    
    outliers_summary = []
    
    for col in numeric_cols[:6]:  # Limiter aux 6 premi√®res colonnes num√©riques
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        
        outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
        outlier_count = len(outliers)
        outlier_percent = (outlier_count / len(df)) * 100
        
        outliers_summary.append({
            'Variable': col,
            'Nombre': outlier_count,
            'Pourcentage': outlier_percent,
            'Borne inf': lower_bound,
            'Borne sup': upper_bound
        })
        
    outliers_df = pd.DataFrame(outliers_summary)
    display(outliers_df)

# %% [markdown]
# ## 7. R√©sum√© et recommandations

# %%
if df is not None:
    print("üìã R√âSUM√â DE L'EXPLORATION")
    print("=" * 50)
    print(f"üìä Taille du dataset: {df.shape[0]:,} lignes √ó {df.shape[1]} colonnes")
    print(f"üî¢ Variables num√©riques: {len(numeric_cols)}")
    print(f"üìù Variables cat√©gorielles: {len(categorical_cols)}")
    
    missing_total = df.isnull().sum().sum()
    print(f"‚ùì Valeurs manquantes: {missing_total:,} ({(missing_total/(df.size))*100:.2f}%)")
    
    if price_cols:
        target_col = price_cols[0]
        print(f"üéØ Variable cible: {target_col}")
        print(f"üí∞ Prix moyen: ${df[target_col].mean():,.2f}")
    
    print("\nüîß RECOMMANDATIONS POUR LE PREPROCESSING:")
    print("=" * 50)
    
    # Recommandations bas√©es sur l'exploration
    if missing_total > 0:
        print("‚Ä¢ G√©rer les valeurs manquantes (imputation ou suppression)")
    
    if len(categorical_cols) > 0:
        print("‚Ä¢ Encoder les variables cat√©gorielles (One-Hot, Label)")
        
    if price_cols:
        target_col = price_cols[0]
        skewness = df[target_col].skew()
        if abs(skewness) > 1:
            print(f"‚Ä¢ Transformer la variable cible (skewness: {skewness:.2f})")
    
    print("‚Ä¢ D√©tecter et traiter les valeurs aberrantes")
    print("‚Ä¢ Normaliser/standardiser les variables num√©riques")
    print("‚Ä¢ Cr√©er de nouvelles features si pertinent")

# %%
print("‚úÖ Exploration termin√©e!")
print("üìÅ Prochaine √©tape: Preprocessing des donn√©es")
```
